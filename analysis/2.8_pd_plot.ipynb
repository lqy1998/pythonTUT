```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.Series(np.random.randn(1000))
data = data.cumsum()
data.head(3)
```

```python
data.plot()
plt.show()
```

```python
data = pd.DataFrame(np.random.randn(1000,4), columns=list('ABCD'))
data = data.cumsum()
data.head(5)
```

```python
data.plot()
plt.show()
```

```python
bx = data.plot.scatter(x='A', y='B', color='LightBlue', label='Class 1')
data.plot.scatter(x='A', y='C', color='DarkGreen', label='Class 2', ax=bx)
plt.show()
```
