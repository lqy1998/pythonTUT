```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
plt.figure(figsize=(6, 4))
plt.subplot(2, 2, 1)
plt.plot([0, 1], [0, 1])

plt.subplot(222)
plt.plot([0, 1], [0, 2])

plt.subplot(223)
plt.plot([0, 1], [0, 3])

# plt.subplot(224)
# plt.plot([0, 1], [0, 4])

# automatically adjust the space among subplots
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.subplot(2, 1, 1)
plt.plot([0, 1], [0, 1])

plt.subplot(234)
plt.plot([0, 1], [0, 2])

plt.subplot(235)
plt.plot([0, 1], [0, 3])

plt.subplot(236)
plt.plot([0, 1], [0, 4])

plt.tight_layout()
plt.show()
```

```python
# method 1: easy to define structure
####################################
fig, ((ax11, ax12), (ax13, ax14)) = plt.subplots(2, 2, sharex=True, sharey=True)
ax11.scatter([1,2], [1,2])

fig.canvas.set_window_title('My Window Title') 
plt.tight_layout()
plt.show()
```

```python
# method 2: subplot2grid
##########################
plt.figure()

# 3r3c, start from(0, 0)
ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=3)
ax1.plot([1, 2], [1, 2])
ax1.set_title('ax1_title')

ax2 = plt.subplot2grid((3, 3), (1, 0), colspan=2)
plt.title('aaaaa')

ax3 = plt.subplot2grid((3, 3), (1, 2), rowspan=2)

ax4 = plt.subplot2grid((3, 3), (2, 0))
ax4.scatter([1, 2], [2, 2])
ax4.set_xlabel('ax4_x')
ax4.set_ylabel('ax4_y')

ax5 = plt.subplot2grid((3, 3), (2, 1))
plt.xlabel('bbbbb')

plt.tight_layout()
plt.show()
```

```python
# method 3: gridspec   grid specification
#########################
plt.figure('fig')
gs = gridspec.GridSpec(3, 3)

# use index from 0
ax6 = plt.subplot(gs[0, :])
ax7 = plt.subplot(gs[1, :2])
ax8 = plt.subplot(gs[1:, 2])
ax9 = plt.subplot(gs[-1, 0])
ax10 = plt.subplot(gs[2, -2])

plt.tight_layout()
plt.show()
```
